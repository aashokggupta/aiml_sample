```python
from langchain_core.prompts import PromptTemplate
from langchain_classic.chains import LLMChain, SequentialChain
from langchain_classic.memory import ConversationBufferMemory
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
import os

llm = ChatGoogleGenerativeAI(model='gemini-2.5-flash', google_api_key=os.getenv('GOOGLE_API_KEY'))
```

```python
title_prompt = PromptTemplate(
    input_variables=['topic'],
    template='write me a catchy blog title about {topic}'
)

title_chain=LLMChain(
    llm=llm,
    prompt=title_prompt,
    output_key='title',
    verbose=True
)
```

```python
outline_prompt=PromptTemplate(
    input_variables=['title'],
    template='write me a blog outline for the title {title}'
)

outline_chain=LLMChain(
    llm=llm,
    prompt=outline_prompt,
    output_key='outline'
)
```

```python
blog_prompt= PromptTemplate(
    input_variables=['title', 'outline'],
    template='write me a blog post for the title {title} and outline {outline} in 100 words'
)

blog_chain=LLMChain(
    llm=llm,
    prompt=blog_prompt,
    output_key='blog'
)
```

```python
summary_prompt=PromptTemplate(
    input_variables=['blog'],
    template='write me a summary of the blog post {blog} 25 words'
)

summary_chain=LLMChain(
    llm=llm,
    prompt=summary_prompt,
    output_key='summary'
)
```

```python
overall_chain=SequentialChain(
    chains=[title_chain, outline_chain, blog_chain, summary_chain],
    input_variables=['topic'],
    output_variables=['title', 'outline', 'blog', 'summary'],
    verbose=True
)
```

```python
result=overall_chain.invoke({"topic":"AI driven cybersecurity in banking domain"})
```

Scenario 1 - Marketing Campaign Generation A company is launching a new AI powered fitness smartwatch. The marketing team wants to create a complete campaign, including audience, identification, messaging, social media content and email outreach

Scenario 2 - Banking Loan Approval Assistant A bank receives a home loan application and wants to automate preliminary loan assessment.

### Scenario 1: Marketing Campaign Generation for an AI-Powered Fitness Smartwatch

```python
# Define prompt for Audience Identification
audience_prompt = PromptTemplate(
    input_variables=['product_description'],
    template='Identify the target audience for an AI powered fitness smartwatch given the product description: {product_description}'
)

audience_chain = LLMChain(
    llm=llm,
    prompt=audience_prompt,
    output_key='audience',
    verbose=True
)
```

```python
# Define prompt for Messaging
messaging_prompt = PromptTemplate(
    input_variables=['product_description', 'audience'],
    template='Based on the product description: {product_description} and the identified audience: {audience}, create compelling marketing messages.'
)

messaging_chain = LLMChain(
    llm=llm,
    prompt=messaging_prompt,
    output_key='messaging',
    verbose=True
)
```

```python
# Define prompt for Social Media Content
social_media_prompt = PromptTemplate(
    input_variables=['product_description', 'messaging'],
    template='Given the product description: {product_description} and key marketing messages: {messaging}, generate engaging social media content (e.g., a few posts for Instagram/Facebook with relevant hashtags).'
)

social_media_chain = LLMChain(
    llm=llm,
    prompt=social_media_prompt,
    output_key='social_media',
    verbose=True
)
```

```python
# Define prompt for Email Outreach
email_outreach_prompt = PromptTemplate(
    input_variables=['product_description', 'messaging', 'audience'],
    template='For the product: {product_description}, using marketing messages: {messaging}, and targeting audience: {audience}, draft a concise email outreach copy for a product launch.'
)

email_outreach_chain = LLMChain(
    llm=llm,
    prompt=email_outreach_prompt,
    output_key='email_outreach',
    verbose=True
)
```

```python
# Create the overall SequentialChain for the marketing campaign
marketing_campaign_chain = SequentialChain(
    chains=[audience_chain, messaging_chain, social_media_chain, email_outreach_chain],
    input_variables=['product_description'],
    output_variables=['audience', 'messaging', 'social_media', 'email_outreach'],
    verbose=True
)
```

```python
# Define the product description
product_desc = "A cutting-edge AI-powered fitness smartwatch that tracks advanced health metrics, provides personalized workout recommendations, and monitors sleep patterns with real-time feedback. It features a sleek design, long-lasting battery, and seamless integration with popular fitness apps."

# Run the marketing campaign chain
marketing_results = marketing_campaign_chain.invoke({"product_description": product_desc})
```

#### Generated Marketing Campaign Components:

```python
print("\n--- Target Audience ---\n", marketing_results['audience'])
print("\n--- Marketing Messages ---\n", marketing_results['messaging'])
print("\n--- Social Media Content ---\n", marketing_results['social_media'])
print("\n--- Email Outreach ---\n", marketing_results['email_outreach'])
```

### Scenario 2: Banking Loan Approval Assistant

```python
# Define prompt for Extracting Applicant Information
extraction_prompt = PromptTemplate(
    input_variables=['loan_application'],
    template='Extract the following key information from the loan application: applicant name, requested loan amount, income, credit score, employment status, and any red flags. Format the output as a clear summary. Loan application: {loan_application}'
)

extraction_chain = LLMChain(
    llm=llm,
    prompt=extraction_prompt,
    output_key='applicant_info',
    verbose=True
)
```

```python
# Define prompt for Creditworthiness Assessment
credit_prompt = PromptTemplate(
    input_variables=['applicant_info'],
    template="""Based on the following applicant information: {applicant_info}, assess the applicant's creditworthiness. Consider factors like income vs. loan amount, credit score, and employment stability. Provide a brief assessment (e.g., 'High', 'Medium', 'Low')."""
)

credit_chain = LLMChain(
    llm=llm,
    prompt=credit_prompt,
    output_key='credit_assessment',
    verbose=True
)
```

```python
# Define prompt for Risk Assessment
risk_prompt = PromptTemplate(
    input_variables=['applicant_info', 'credit_assessment'],
    template='Given the applicant information: {applicant_info} and creditworthiness: {credit_assessment}, identify any potential risks associated with approving this loan. (e.g., high debt-to-income ratio, unstable employment, low credit score).'
)

risk_chain = LLMChain(
    llm=llm,
    prompt=risk_prompt,
    output_key='risk_assessment',
    verbose=True
)
```

```python
# Define prompt for Preliminary Recommendation
recommendation_prompt = PromptTemplate(
    input_variables=['applicant_info', 'credit_assessment', 'risk_assessment'],
    template="""Based on the applicant information: {applicant_info}, creditworthiness: {credit_assessment}, and identified risks: {risk_assessment}, provide a preliminary loan approval recommendation (e.g., 'Recommend Approval', 'Recommend Rejection', 'Further Review Required') with a brief justification."""
)

recommendation_chain = LLMChain(
    llm=llm,
    prompt=recommendation_prompt,
    output_key='loan_recommendation',
    verbose=True
)
```

```python
# Create the overall SequentialChain for the loan approval process
loan_approval_chain = SequentialChain(
    chains=[extraction_chain, credit_chain, risk_chain, recommendation_chain],
    input_variables=['loan_application'],
    output_variables=['applicant_info', 'credit_assessment', 'risk_assessment', 'loan_recommendation'],
    verbose=True
)
```

```python
# Example Loan Application
example_application = """
Applicant Name: John Doe
Requested Loan Amount: $300,000
Income: $90,000 per year
Credit Score: 720
Employment Status: Employed full-time for 5 years at Tech Solutions Inc.
Previous Loans: Auto loan (paid off), Student loan (current, $20,000 remaining)
Red Flags: None
"""

# Run the loan approval chain
loan_results = loan_approval_chain.invoke({"loan_application": example_application})
```

#### Generated Loan Assessment Components:

```python
print("\n--- Applicant Information ---\n", loan_results['applicant_info'])
print("\n--- Creditworthiness Assessment ---\n", loan_results['credit_assessment'])
print("\n--- Risk Assessment ---\n", loan_results['risk_assessment'])
print("\n--- Loan Recommendation ---\n", loan_results['loan_recommendation'])
```
