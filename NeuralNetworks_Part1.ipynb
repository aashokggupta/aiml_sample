Building a Binary Classifier using Neural Network

We are classifying loan approval or disapproval case using input features like:

1. Income of a customer
2. Credit score of a customer
3. Years of experience of a customer

```python
import numpy as np
```

```python
#input features - input to the neural network
X=np.array([
    [30000,550,1],
    [80000,720,6],
    [50000,600,3],
    [90000,750,8],
    [35000,580,2]
])

#actual output label - 0 (loan is not approved), 1(loan is approved)
y=np.array([0,1,0,1,0]) #correct answers for learning
```

image.png

```python
#building neural network architecture
import tensorflow as tf
from tensorflow.keras import Sequential, layers
```

```python
model=Sequential([
    layers.Dense(3,activation='relu',input_shape=(3,)), #input layer
    layers.Dense(3,activation='relu'), #1st hidden layer
    layers.Dense(1,activation='sigmoid') #output layer
])

model.summary()
#None because the batch size is not known to us which means we have not mentioned how many records will be passed at once to the NN
```

Formula for calculating Parameters:
- (No. of Input features * Number of Neurons) + Number of Weights

E.g. In the first layer, the total number of Parameters are 12 because:
- (3*3)+3 = 12 params

E.g. In the output layer, the total number of Parameters are 4 because:
- (3*1)+1 = 4 params
