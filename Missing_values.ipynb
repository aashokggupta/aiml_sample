```python
import numpy as np

mydict = {
    "income": (50, 25, 15, 10, np.nan, np.nan, 20, 10, np.nan, 999),
    "Age": (19, 20, 30, np.nan, np.nan, 25, 20, 52, 46, 9999)
}

mydict
```

```python
mydata=pd.DataFrame(mydict)
mydata
```

```python
#replace bogus values such as 9999with nan
mydata = mydata.replace((9999,999), np.nan)
mydata
```

```python
#Display values which are "Null' with 'True'
mydata.isnull()
```

```python
#Drop any rows where there are any columns missing values
new=mydata.dropna(how='any')
new
```

```python
#Drop any rows where there are all columns missing values
new=mydata.dropna(how='all')
new
```

```python
#Replace missing values with mean of respective columns
new=np.round(mydata.fillna(mydata.mean()),1)
new
```

```python
#Replace missing values with median of respective columns
new=np.round(mydata.fillna(mydata.mean()),1)
new
```

```python
#Replace missing values with any numbers of your choice
new=np.round(mydata.fillna({'Age':30,'income':15}))
new
```

Pandas join Table

```python
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

# Create DataFrames
married = pd.DataFrame({
    "emp_id": (1,2,3,4,5,6,7,8),
    "Married": ['John','James','Harper','Amy','Mia','Liz','Benjamin','Ava']
})

homeowners = pd.DataFrame({
    "emp_id": (9,10,11,12,5,6,7,8),
    "Homeowners": ['Sophia','William','Jacob','Michael','Mia','Liz','Benjamin','Ava']
})

# Sorting
married_sorted = married.sort_index(axis=1, ascending=False)
homeowners_sorted = homeowners.sort_values(by='emp_id', ascending=True)

# Print outputs
print(married_sorted)
print()
print(homeowners_sorted)
```

```python
# Print outputs
married_sorted
homeowners_sorted
```

```python
#Find employees who are married and homemakers
consolidated = pd.merge(married, homeowners, on='emp_id', how='inner')
consolidated.sort_index(axis=1, ascending=False)
```

```python
#Find employees who are married and may or maynot be homemakers
consolidated = pd.merge(married, homeowners, on='emp_id', how='left')
consolidated.sort_index(axis=1, ascending=False)
```

```python
#Find employees who are homemakersand may or maynot be married
consolidated = pd.merge(married, homeowners, on='emp_id', how='right')
consolidated.sort_index(axis=1, ascending=False)
```

```python
#Find employees who are homemakersand or married
consolidated = pd.merge(married, homeowners, on='emp_id', how='outer')
consolidated.sort_index(axis=1, ascending=False)
```

Pandas Pivot table

```python
#Example 1: Load dataset
import pandas as pd
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
#Example 2: Average total bill by day and time
pd.pivot_table(
    tips,
    values="tip",
    index="day",
    aggfunc="mean"
)
```

```python
#Example 3: Multiple aggregations
pd.pivot_table(
    tips,
    values="total_bill",
    index="day",
    columns="time",
    aggfunc="mean"
)
```

```python
##Example 3: Multiple aggregations
pd.pivot_table(
    tips,
    values="tip",
    index="day",
    aggfunc=["mean", "sum", "count"]
)
```

```python
#Example 4: Pivot table with filters
pd.pivot_table(
    tips,
    values="tip",
    index="day",
    columns="sex",
    aggfunc="mean",
    margins=True
)
```
