```python
#import pandas and scikit package
import pandas as pd
import sklearn.datasets as datas
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
#Load the iris dataset
irisdata = datas.load_iris()
irisdata
```

```python
#check datatype
type(irisdata)
```

```python
#examine the keys present in the dataset
irisdata.keys()
```

```python
irisdata.DESCR
```

```python
irisdata.target
```

```python
#collect all the indepent variables and store it as an object irisIndData
irisIndData = irisdata.data
```

```python
#convert the object into a pandas dataframe
irisIndData=pd.DataFrame(irisdata.data)
irisIndData.head(5)
```

```python
#Display the feature names in the dataset
irisdata.feature_names
```

```python
#create an list with all the column names
irisIndData.columns = ['SepalLength', 'SepalWidth','PetalLength','PetalWidth']
irisIndData.head()
```

```python
#split the dataset-80% to training dataset and 20% to test dataset
irisIndTrain, irisIndTest, irisDepTrain, irisDepTest  = train_test_split(irisIndData,irisdata.target,test_size=0.2, random_state=0)
```

```python
#Instantiate the decision tree model
dt  = tree.DecisionTreeClassifier(max_depth=4, criterion='entropy')
```

```python
#Build model
modl = dt.fit(irisIndTrain,irisDepTrain)
```

```python
#Obtain the predictions for the model
predictedData = modl.predict(irisIndTest)
predictedData
```

```python
#Import graphviz package
import graphviz
```

```python
#Visualize the decision tree
dot_data=tree.export_graphviz(dt,out_file=None,feature_names=irisdata.feature_names,
                              class_names=irisdata.target_names,filled=True,rounded=True,
                              special_characters=True)
```

```python
#Display the Decision Tree
graph=graphviz.Source(dot_data)
graph
```
