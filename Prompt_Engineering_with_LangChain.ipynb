```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI #helps in configuring the required LLM
from langchain_core.prompts import ChatPromptTemplate, FewShotChatMessagePromptTemplate
```

```python
os.environ["GOOGLE_API_KEY"]="API" #adding the Google Gemini LLM's API
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-2.5-flash") #initializing the Gemini llm
```

Incase of LLMs we have
- System Message: Persona of the LLM
- User Message: User Query

```python
#zero-shot prompting - Product Positioning
zero_shot_prompt=ChatPromptTemplate.from_messages([
    ("system","You are a Senior Marketing Strategist"), #persona attached to the LLM
    ("human","""Create a concise product positioning statement for an AI powered resume-screening platform.
    Keep it as simple as possible in terms of language usage so that even a small kid can understand the rationale behind this product""") #user query
])
```

```python
response=llm.invoke(zero_shot_prompt.format_messages())
print(response)
```

```python
#Chain of Thought prompting - Loan Eligibility Assessment

prompt=ChatPromptTemplate.from_messages([
    ("system","You are a Credit Risk Analyst at a Commercial Bank"),
    ("human","""
    Assess the loan eligibility using the following structured evaluation method provided below:

    Applicant Details:
    1. Credit Score: 700
    2. Monthly Income:2,50,000 INR
    3. Existing EMI Obligations: 20,000 INR
    4. Loan Amount Requested: 10,00,000 INR

    Evaluation Steps:
    1. Creditworthiness Assessment
    2. Affordability Analysis
    3. Risk consideration
    4. Final eligibility decision

    Output format:
    - Clear section for headers
    - Bullet points under each section
    - Final decision clearly stated with reason"""

    )
])
```

```python
response=llm.invoke(prompt.format_messages())
```
