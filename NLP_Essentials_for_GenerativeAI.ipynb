We are building a NLP Model where it converts the Text Sentence:
- into constituting Tokens
- removes stop-words from the Tokens
- performs Stemming and checks the redundancies
- performs Lemmatization and check the improved quality of output generated
- converts the output generated to corresponding Vectorial representation

```python
#importing dependencies
import nltk #natural language toolkit
from nltk.tokenize import word_tokenize #for word level tokenization
from nltk.corpus import stopwords, wordnet
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
from gensim.models import Word2Vec #gensim is a pretrained embedding model used for converting text to vectors
```

```python
nltk.download("stopwords") #download the most commonly used words in any given language (a, the, he, she etc.)
nltk.download("wordnet") # dictionary of any given language where we have both - the word and synonymns of the word
nltk.download("punkt_tab") #pre-trained unsupervised tokenizer
nltk.download("omw-1.4") #Opensource Multi-Lingual WordNet
```

```python
sentences=["The learners are learning about Natural Language Processing in Artificial Intelligence"]
```

```python
#Tokenization
tokenized_sentence=[word_tokenize(sentence.lower())for sentence in sentences]
print("Tokenized",tokenized_sentence)
```

```python
#Stopword Removal
stop_words=set(stopwords.words('english'))
print(stop_words)
```

```python
len(stop_words) #there are 198 words most commonly used in English Language that acts as connectors
```

```python
new_filtered_sentences=[
    [word for word in sentence if word.isalnum() and word not in stop_words]
    for sentence in tokenized_sentence
]
print("after stopwords removal",new_filtered_sentences)
```

```python
print("Tokenized",tokenized_sentence)
```

```python
#stemming
stemmer=PorterStemmer()
stemmed_sentences=[
    [stemmer.stem(word)for word in sentence]
    for sentence in new_filtered_sentences
]

print("stemmed sentences",stemmed_sentences)
```

```python
print("after stopwords removal",new_filtered_sentences)
```

```python
#lemmatization
lemmatizer=WordNetLemmatizer()
new_lemmatized_sentences=[
    [lemmatizer.lemmatize(word) for word in sentence]
    for sentence in new_filtered_sentences
]

print("lemmatized sentences",new_lemmatized_sentences)
```

```python
print("after stopwords removal",new_filtered_sentences)
```

```python
print("stemmed sentences",stemmed_sentences)
```

```python
#Vectorial representation using Word2Vec
model=Word2Vec(sentences=new_lemmatized_sentences,vector_size=1,window=1,min_count=1,workers=4)
print("Vectorial representation of the word 'artificial' is",model.wv['artificial'])
```

- Vector size is the dimention of the output vector generated
- Window is the number of words that we consider before and after the query word to understand the context
- Workers will help in processing the output (Number of cores)

Task:

- Use a sentence like - Artificial Intelligence is one of the most important subjects in Computer Science Engineering

- Use a sentence like - The beauty specialist was so novice that the makeover done was looking very artificial

```python
sentence1='Artificial Intelligence is one of the most important subjects in Computer Science Engineering'
tokenized_sentence1 = word_tokenize(sentence1.lower())
model1=Word2Vec(sentences=[tokenized_sentence1],vector_size=1,window=1,min_count=1,workers=4)
print("Vectorial representation of the word 'artificial' is",model1.wv['artificial'])

sentence2='The beauty specialist was so novice that the makeover done was looking very artificial'
tokenized_sentence2 = word_tokenize(sentence2.lower())
model2=Word2Vec(sentences=[tokenized_sentence2],vector_size=1,window=1,min_count=1,workers=4)
print("Vectorial representation of the word 'artificial' is",model2.wv['artificial'])
```
