We are utilizing Transformer as a pretrained model for:
- Text Classification
- Text Generation

```python
text1='''Extremely disappointed with the recent purchase of IPhone 17 from Apple.
This device constantly lags and the battery is not upto the mark. It barely lasts through the day.'''

text2='''I recently purchased an IPhone from Apple and it has been an absolute delight.
This device runs smoothly and the battery life is impressive. The price, though high, is justified by the
excellent performance and top notch customer support'''
```

```python
from transformers import pipeline #pipelines in transfomers are different for different usecases
#e.g. pipeline of text classification is very different from text generation

classifier=pipeline('text-classification') #we need to explicitly specify

import pandas as pd

output1=classifier(text1)
```

```python
pd.DataFrame(output1)
```

```python
output2=classifier(text2)
```

```python
pd.DataFrame(output2)
```

```python
generator=pipeline('text-generation')
response="Dear valued customer, thanks for writing!I am sorry to hear your experience with us"
prompt=text1+'\n\Customer Service Response:\n'+response

output=generator(prompt,max_length=100)
#along with the response, Transformer will add 100 tokens to complete the message
```

```python
print(output)
```
