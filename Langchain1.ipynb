Building a Sequential Workflow using LangChain

User shares a Topic
- Chain 1 is responsible for creating a 'Title' out of the 'Topic' shared by the User
- Chain 2 is responsible for creating an 'Outline' from the 'Title'
- Chain 3 is responsible for creating a 'Blog draft' from the 'Outline'
- Chain 4 is responsible for creating a 'Summary' from the 'Blog draft'

```python
from langchain_core.prompts import PromptTemplate
from langchain_classic.chains import LLMChain, SequentialChain
from langchain_classic.memory import ConversationBufferMemory
```

```python
llm=ChatGoogleGenerativeAI(model='gemini-2.5-flash')
```

```python
#Chain 1 - Generating a title for the Blog
title_prompt=PromptTemplate(
    input_variables=['topic'], #this topic comes from the user
    template='Create a catchy blog title for the given : {topic}'
)

title_chain=LLMChain(
    llm=llm, #ChatGoogleGenerativeAI(model='gemini-2.5-flash')
    prompt=title_prompt,
    output_key='title',
)
```

```python
#Chain 2 - Generating an outline from the title generated
outline_prompt=PromptTemplate(
    input_variables=['title'], #has already been generated as a result of execution of the previous chain
    template='Provide a structured outline for the blog with a title : {title}'
)

outline_chain=LLMChain(
    llm=llm,
    prompt=outline_prompt,
    output_key='outline'
)
```

```python
#Chain 3 - Expanding the outline into a blog
blog_prompt=PromptTemplate(
    input_variables=['outline'],
    template='Write a detailed blog based on the following outline: {outline}'
)

blog_chain=LLMChain(
    llm=llm,
    prompt=blog_prompt,
    output_key='blog'
)
```

```python
#Chain 4 - Summarize the blog
summary_prompt=PromptTemplate(
    input_variables=['blog'],
    template='Summarize the following blog in 100 words or less: {blog}'
)

summary_chain=LLMChain(
    llm=llm,
    prompt=summary_prompt,
    output_key='summary'
)
```

```python
#Combining into Sequential Chain
overall_chain=SequentialChain(
    chains=[title_chain, outline_chain, blog_chain, summary_chain],
    input_variables=['topic'], #from the user
    output_variables=['title', 'outline', 'blog', 'summary'],
)
```

```python
result=overall_chain({"topic":"AI driven Cybersecurity in Banking Domain"})
```
